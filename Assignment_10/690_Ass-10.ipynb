here using groupby to get mean of life expectancy based on year

```python
df.groupby('year')['lifeExp'].mean()
```

```python
df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean()
```

```python
# here we use weather csv file
weather = pd.read_csv('pydatadc_2018-tidy-master/pydatadc_2018-tidy-master/data/weather.csv')
```

```python
weather
```

melting weather data 

```python
# can use pd.melt, or weather.melt
weather_melt = weather.melt(
    id_vars=['id', 'year', 'month', 'element'],
    var_name='day',
    value_name='temp'
)
```

```python
weather_melt.head()
```

here pivot_table used for dataframe to store like a work sheet format

```python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp'
)
```

if we want a regular dataframe use reset_index then it will flatten everything

```python
weather_tidy.reset_index()
```
