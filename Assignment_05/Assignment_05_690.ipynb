# Definitions for the above 9 variables

**UNITID** - Unit ID for institution, it is a integer data type

**INSTNM** - Institution names, it is a string data type

**CITY** - City names, it is a string data type

**STABBR** - State postcode, it is a string data type

**ZIP** - Zip codes, it is a string data type

**ADM_RATE** - Admission rate, it is a float data type. 
Colleges report to IPEDS their Fall admissions rate, defined as the number of admitted undergraduates divided by the number of
undergraduates who applied. For institutions with multiple branches, ADM_RATE includes the admissions rate at each campus.
(Fall 2017, reported in IPEDS DCY2017-18)

**UGDS** - Number of Undergraduate Students, it is a integer data type. 
This element (UGDS) includes the number of degree/certificate-seeking undergraduates enrolled in the fall, as reported in the IPEDS Fall Enrollment component.
(Fall 2017, reported in IPEDS DCY2017-18)

**TUITIONFEE_IN** - Tuition and Fees,  the cost data include the tuition and required fees of the institution. They are provided for in-state students (TUITIONFEE_IN).
(AcadYr 2017-18, reported in IPEDS DCY2017-18)

**MN_EARN_WNE_P10** - Mean and Median Earnings, it is a integer data type.
Mean (MN_EARN_WNE_P*) and median (MD_EARN_WNE_P*) earnings are for the institutional aggregate of all federally aided students who enroll in an institution each year and who are employed but not enrolled.
