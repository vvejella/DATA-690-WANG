# removing last element

```python
df['Census Year'] = df['Census Year'].iloc[:-1]
```

# removing commas

```python
df['Total Population'] = df['Total Population'].str.replace(',', '')
```

#  removing dollar and replacing billion

```python
df['Census Cost'] = df['Census Cost'].str.replace('$', '')
```

```python
df['Census Cost'] = df['Census Cost'].str.replace(',', '')
```

```python
df['Census Cost'] = df['Census Cost'].str.replace(' Billion','0000000000')
```

```python
df['Average Cost Per Person'] = df['Average Cost Per Person'].str.replace('$', '')
```

```python
xy = df['Average Cost Per Person'].values.tolist()
```

```python
lis = []
for x in xy:
    if x.endswith(' cents'):
        y = x.replace(' cents','')
        lis.append(float(y)/100)
    else:
        lis.append(x)
print(lis)   
```

```python
df['avg_list'] = pd.DataFrame(lis)
```

```python
df = df.drop(['Average Cost Per Person'],axis = 1)
```
